KNN Algorithm Summary
Distance Measurement: KNN starts by calculating the distance from a query point to other points in the dataset. Although the Euclidean distance is common, it’s sometimes simplified by omitting the square root operation, as the relative distances remain the same.

Sorting by Distance: Instead of iterating through the entire dataset, sorting the points by their distance to the query point is more efficient. The algorithm then selects the nearest k points. The computational complexity for sorting is 
O(NlogN)
O(NlogN).

Optimization Using Quad Trees: For larger datasets, quad trees help by spatially indexing the data, enabling quicker searches by segmenting the space into quadrants. This way, not all points need to be examined—only those in quadrants that intersect with the query’s range.

Range Query: When the search radius isn’t predefined, KNN starts with an arbitrarily large radius and narrows it down as closer points are found. The radius becomes the distance to the k-th nearest point found so far, allowing for efficient pruning of further points.

Handling Leaf Nodes: When reaching a leaf node in the tree structure, the algorithm checks the points within. If fewer than k points have been found, it adds new ones. If more, it only considers those closer than the furthest point in the current nearest set.
